```python
import uuid
docs = []
metadatas = []
# embeddings = []
ids = []
for i, row in icd_data.iterrows():
    # Combine for semantic search
    parts = [row['code_description']]

    if not pd.isna(row['clinical_information']) and row['clinical_information'] != '':
        parts.append(f"Clinical Information: {row['clinical_information']}")

    if not pd.isna(row['applicable_to']) and row['applicable_to'] != '':
        parts.append(f"Also known as: {row['applicable_to']}")

    if not pd.isna(row['approximate_synonyms']) and row['approximate_synonyms'] != '':
        parts.append(f"Synonyms: {row['approximate_synonyms']}")

    # print(parts)
    combined_text = ".\n".join(parts)  # Natural sentence boundary
    
    # Also keep synonym tags for exact keyword filtering if needed
    metadata = {
        "code": row['code_y'],
        "billable": row['billable'],
        "href": row['href'],
        "description": row['code_description']
    }
    
    # Add boolean flags for filtering
    if not pd.isna(row['approximate_synonyms']):
        for syn in row['approximate_synonyms'].split('\n'):
            if syn.strip():
                metadata[f"{syn.strip().lower()}"] = True
    
    docs.append(combined_text)
    ids.append(row['code_y'])
    metadatas.append(metadata)


```

```python
chroma_collection.upsert(
    ids=ids,
    documents=docs,
    metadatas=metadatas
)
```
